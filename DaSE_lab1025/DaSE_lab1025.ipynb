根据daily_KP_SUN_2020.csv绘制日照时间图

```python
import numpy as np
import matplotlib.pyplot as plt
F = open('./daily_KP_SUN_2020.csv', encoding='UTF-8')
Data = F.readlines()
F.close()
monthSum = np.zeros(9)
monthDay = np.zeros(9)
monthAve = np.zeros(9)
for i in range(3, 277):
    temp = Data[i].split(',')
    monthSum[int(temp[1])-1] += float(temp[3])
    monthDay[int(temp[1])-1] += 1
    monthAve[int(temp[1])-1] = monthSum[int(temp[1])-1] / monthDay[int(temp[1])-1]
x = np.arange(1, 10)
plt.subplot(121)
plt.title("sum of daily sun time")
plt.bar(x, monthSum)
plt.subplot(122)
plt.title("average of daily sun time")
plt.bar(x, monthAve)
```

根据iris.csv寻找两个变量确定鸢尾花的种类(red=setosa,green=versicolor,blue=virginica)

```python
import numpy as np
import matplotlib.pyplot as plt
F = open('iris.csv', encoding='UTF-8')
Data = F.readlines()
F.close()
SL = []
SW = []
PL = []
PW = []
color = []
for i in range(1, 151):
    temp = Data[i].split(',')
    SL.append(float(temp[0]))
    SW.append(float(temp[1]))
    PL.append(float(temp[2]))
    PW.append(float(temp[3]))
    if temp[4] == "setosa\n":
        color.append("red")
    elif temp[4] == "virginica\n":
        color.append("blue")
    else:
        color.append("green")
plt.subplot(231)
plt.scatter(SL, SW, s=10, c=color, alpha=0.8)
plt.subplot(232)
plt.scatter(SL, PL, s=10, c=color, alpha=0.8)
plt.subplot(233)
plt.scatter(SL, PW, s=10, c=color, alpha=0.8)
plt.subplot(234)
plt.scatter(SW, PL, s=10, c=color, alpha=0.8)
plt.subplot(235)
plt.scatter(SW, PW, s=10, c=color, alpha=0.8)
plt.subplot(236)
plt.scatter(PL, PW, s=10, c=color, alpha=0.8)
```
