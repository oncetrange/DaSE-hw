1.数据获取

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


data = pd.read_csv('./bike.csv')
data
```

2.数据预处理 I

```python
data = data.drop(columns=['id'])
data
```

3.数据预处理II

```python
data = data.loc[data['city']==1]
data = data.drop(columns=['city'])
data
```

4.数据预处理III

```python
data.loc[data['hour']>=19, 'hour']=0
data.loc[data['hour']<=5, 'hour']=0
data.loc[data['hour']!=0, 'hour']=1
data
```

5.数据预处理IV

```python
y = data['y'].to_numpy()
y = y.reshape(len(y), 1)
data = data.drop(columns=['y'])
data
```

6.数据预处理V

```python
data = data.to_numpy()
print(type(data))
```

7.数据集划分

```python
x_train,x_test,y_train,y_test = train_test_split(data, y, test_size=0.2)
```

8.数据预处理VI

```python
scale = MinMaxScaler()
x_train = scale.fit_transform(x_train)
x_test = scale.fit_transform(x_test)
y_train = scale.fit_transform(y_train)
y_test = scale.fit_transform(y_test)
print(f'x_train:{x_train}')
print(f'y_train:{y_train}')
print(f'x_test:{x_test}')
print(f'y_test:{y_test}')
```

9.模型构建

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

10.模型测试

```python
y_predict = model.predict(x_test)
```

11.模型评估

```python
print(mean_squared_error(y_test,y_predict)**0.5)
```
